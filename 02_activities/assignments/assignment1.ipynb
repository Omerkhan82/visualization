```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Lesson 1: Getting Started with Matplotlib
# Creating a simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot with Matplotlib')
plt.legend()
plt.show()
```

```python
# Lesson 2: Choosing the Right Visualization
# Bar chart example
data = {'Category': ['A', 'B', 'C', 'D'], 'Values': [10, 20, 15, 25]}
df = pd.DataFrame(data)

plt.figure(figsize=(6, 4))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
# Lesson 3: Subplots and Combining Visualizations
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Line plot
ax[0].plot(x, y, color='red')
ax[0].set_title('Sine Wave')

# Bar chart
ax[1].bar(df['Category'], df['Values'], color='green')
ax[1].set_title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
# Lesson 4: Accessible Data Visualization
plt.figure(figsize=(6, 4))
plt.bar(df['Category'], df['Values'], color=['#377eb8', '#ff7f00', '#4daf4a', '#f781bf'])
plt.xlabel('Categories', fontsize=12, fontweight='bold')
plt.ylabel('Values', fontsize=12, fontweight='bold')
plt.title('Improved Accessibility in Data Visualization')
plt.show()
```

```python
# Lesson 5: Data Visualization as Advocacy
# Scatter plot with regression line using Seaborn
np.random.seed(42)
data = {'X': np.random.rand(50) * 10, 'Y': np.random.rand(50) * 50}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.regplot(x='X', y='Y', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Using Data Visualization to Show Trends')
plt.show()
```

```python
# Lesson 6: Beyond Matplotlib (Exploring Seaborn)
sns.set(style='whitegrid')
df = sns.load_dataset('penguins')

plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='body_mass_g', data=df, palette='coolwarm')
plt.title('Penguin Species vs. Body Mass')
plt.show()
```
